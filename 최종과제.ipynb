```python
import re  # 채점 결과 문자열에서 점수만 추출할 때 사용합니다.
from pathlib import Path  # 경로를 OS와 무관하게 다루기 위해 사용합니다.

from dotenv import load_dotenv  # .env 파일의 OPENAI_API_KEY를 불러옵니다.
from langchain_community.document_loaders import PyPDFLoader  # PDF를 페이지별 Document로 읽습니다.
from langchain_community.vectorstores import FAISS  # 벡터 DB를 저장하고 검색할 때 사용합니다.
from langchain_core.output_parsers import StrOutputParser  # LLM 응답에서 문자열만 꺼냅니다.
from langchain_core.prompts import ChatPromptTemplate  # 질문과 문맥을 넣을 프롬프트 틀입니다.
from langchain_core.runnables import RunnablePassthrough  # 질문 값을 그대로 체인에 전달합니다.
from langchain_openai import ChatOpenAI, OpenAIEmbeddings  # OpenAI LLM과 임베딩 모델을 사용합니다.
from langchain_text_splitters import RecursiveCharacterTextSplitter  # 긴 문서를 검색용 청크로 나눕니다.

from grading_utils import evaluation_items  # 질문, 정답, 출처 페이지를 한 번에 불러옵니다.
from grading_utils import grade_predictions  # 예측 답변을 자동 채점합니다.

load_dotenv(override=True)  # .env 값을 현재 세션에 다시 반영합니다.
```

```python
PDF_PATH = "data/금융투자협회_투자길라잡이_2018.pdf"  # 과제에 사용할 원본 PDF입니다.
DB_PATH = Path("./faiss_index_kofia_guide")  # 1일차·2일차와 공용으로 쓰는 FAISS 경로입니다.

CHUNK_SIZE = 500  # 문서를 한 번에 자를 길이입니다.
CHUNK_OVERLAP = 100  # 청크 사이에 겹치게 둘 길이입니다.
TOP_K = 4  # 질문마다 상위 몇 개 청크를 가져올지 정합니다.
MODEL_NAME = "gpt-5-nano"  # 과제 조건에 맞춰 사용할 모델을 고정합니다.

embedding_model = OpenAIEmbeddings(model="text-embedding-3-small")  # 문서를 벡터로 바꿀 임베딩 모델입니다.
```

```python
if DB_PATH.exists():  # 이미 만든 벡터 DB가 있으면 재사용합니다.
    vectorstore = FAISS.load_local(
        str(DB_PATH),  # FAISS가 저장된 폴더 경로입니다.
        embedding_model,  # 저장할 때 쓴 임베딩 모델과 같아야 합니다.
        allow_dangerous_deserialization=True,  # 최신 LangChain에서 필요한 옵션입니다.
    )
else:  # 처음 실행이라면 PDF를 읽어서 새 벡터 DB를 만듭니다.
    loader = PyPDFLoader(PDF_PATH)  # PDF를 페이지 단위로 읽습니다.
    docs = loader.load()  # 전체 페이지를 Document 리스트로 변환합니다.

    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=CHUNK_SIZE,  # 청크 길이를 고정합니다.
        chunk_overlap=CHUNK_OVERLAP,  # 청크가 약간 겹치게 잘라 문맥 손실을 줄입니다.
        separators=["\n\n", "\n", " ", ""],  # 문단 → 줄바꿈 → 공백 순서로 우선 분할합니다.
    )
    split_docs = text_splitter.split_documents(docs)  # 검색용 청크 리스트를 만듭니다.

    vectorstore = FAISS.from_documents(split_docs, embedding_model)  # 청크를 임베딩해 FAISS 인덱스를 만듭니다.
    vectorstore.save_local(str(DB_PATH))  # 다음 실행부터 바로 쓰도록 로컬에 저장합니다.
```

```python
retriever = vectorstore.as_retriever(search_kwargs={"k": TOP_K})  # 질문마다 상위 TOP_K개 청크를 찾습니다.

def format_docs(docs):  # 검색된 여러 청크를 LLM 입력용 문맥 문자열로 합칩니다.
    lines = []  # 페이지 번호를 붙인 청크 문자열을 모아둘 리스트입니다.
    for doc in docs:  # 검색 결과로 나온 각 Document를 순회합니다.
        page = doc.metadata.get("page")  # 원문 페이지 번호를 가져옵니다.
        prefix = f"[p.{page + 1}] " if page is not None else ""  # 출처 확인이 쉽도록 페이지 번호를 앞에 붙입니다.
        lines.append(prefix + doc.page_content)  # 페이지 정보와 본문을 함께 저장합니다.
    return "\n\n".join(lines)  # 청크 사이를 빈 줄로 띄워 LLM이 읽기 쉬운 문맥으로 만듭니다.

llm = ChatOpenAI(model=MODEL_NAME, temperature=0)  # 답변 생성에 사용할 모델입니다.
prompt = ChatPromptTemplate.from_template('''
다음은 문서에서 검색된 정보입니다.
{context}

규칙:
1) 질문에 대해 위 문맥에 근거해서만 답하세요.
2) 답은 가능한 한 짧고 직접적으로 출력하세요.
3) 설명, 출처, 불릿, 줄바꿈을 붙이지 마세요.
4) 문서에 없으면 반드시 정확히 "없는 정보"라고만 답하세요.

질문: {question}
''')  # 질문과 검색 문맥을 함께 넣는 프롬프트입니다.
parser = StrOutputParser()  # 최종 응답에서 문자열만 꺼냅니다.

chain = (  # 검색 -> 프롬프트 -> LLM -> 문자열 파싱 흐름을 하나로 묶습니다.
    {"context": retriever | format_docs, "question": RunnablePassthrough()}  # 질문을 그대로 넘기고, 검색 문맥은 자동으로 붙입니다.
    | prompt
    | llm
    | parser
)

def extract_scores(llm_output: str):  # 채점 결과 문자열에서 문항별 점수만 뽑아냅니다.
    score_pattern = re.compile(r"\d+번:\s*([01](?:\.5)?)점")  # 1점, 0.5점, 0점 패턴만 찾습니다.
    scores = [float(score) for score in score_pattern.findall(llm_output)]  # 찾은 점수를 숫자 리스트로 변환합니다.
    return sum(scores), scores  # 총점과 문항별 점수를 함께 반환합니다.

```

```python
from grading_utils import evaluation_items

# evaluation_items가 정상적으로 로드되었는지 확인 (에러 발생 시 안내)
try:
    evaluation_items
except NameError:
    raise NameError("grading_utils에서 evaluation_items를 불러오지 못했습니다. 모듈을 확인해주세요.")

# 1. 평가 문항 개수 검증
assert len(evaluation_items) == 20, f"평가 문항 수가 20개가 아닙니다. (현재: {len(evaluation_items)}개)"

questions = [item["question"] for item in evaluation_items]
predicted_answers = []  # 모델 예측 답안 저장 리스트

# 2. 질문 순회 및 답변 생성
for idx, item in enumerate(evaluation_items, start=1):
    question = item["question"]
    
    # LangChain 호출 및 답변 정제
    answer = rag_chain.invoke(
        question,
        config={"callbacks": [langfuse_handler]},  # Langfuse 트레이싱 설정
    ).strip()
    
    predicted_answers.append(answer)
    
    # 실시간 예측 결과 출력
    print(f"\n💬 질문 {idx} [p.{item['source_page']}]: {question}")
    print(f"→ 예측: {answer} | 정답: {item['answer']}")

# 3. 자동 채점 진행
print("\n✅ 자동 채점 결과")
grading_result = grade_predictions(questions, predicted_answers)
print(grading_result)

# 4. 점수 추출 및 최종 결과 출력
total_score, per_item_scores = extract_scores(grading_result)
print(f"\n총점: {total_score}점 / {len(per_item_scores)}문항")

# Langfuse 버퍼 비우기
langfuse.flush()
```
